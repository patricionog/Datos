```python
# Borrar fila con NULL en ID.
Secciones = Secciones.dropna(subset=['office_id'])
```

```python
# Borrar espacios adicionales.

for Column in Secciones.columns:
    Secciones[Column] = Secciones[Column].str.strip().str.replace(r'\s+', ' ', regex=True)
```

```python
# Contar cantidad de secciones.
Count = Secciones['office_id'].value_counts()
```

```python
# Obtener sede y cantidad de secciones.

Offices = []
Sections = []

for Office_ID, Number_Of_Sections in Count.items():
    Offices.append(Office_ID)
    Sections.append(Number_Of_Sections)
```

```python
# Nuevo df con contabilización de secciones.
Secciones_New = pd.DataFrame({'office_id': Offices, 'sections': Sections})

Secciones_New
```
