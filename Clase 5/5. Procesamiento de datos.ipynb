superanSalarioActividad01(empleado_01, 15000) debera devolver:



```python
def superanSalarioActividad01(matriz, valor_umbral):
    matriz_filtrada = []
    for i in range(0,len(matriz)):
        if matriz[i][3] > valor_umbral:
            matriz_filtrada.append(matriz[i])
    return matriz_filtrada

superanSalarioActividad01(empleado_01, 15000)
```
