### __Eligiendo__

```python
# a1. Listar a cada alumno que rindió el Parcial-01 y decir si aprobó o no (se aprueba con nota >=4).
    
Consulta_SQL = """
                SELECT Nombre, 
                    CASE 
                        WHEN Nota >= 4 THEN 'Aprobado'
                        WHEN Nota < 4 THEN 'Desaprobado'
                    END AS Condicion
                FROM Examen
                WHERE Instancia = 'Parcial-01';

               """

df_Resultado = pd.read_sql_query(Consulta_SQL, Connection)

df_Resultado
```

```python
# a2. Modificar la consulta anterior para que informe cuántos estudiantes aprobaron/reprobaron en cada instancia.
    
Consulta_SQL = """
                SELECT Instancia, 
                    CASE 
                        WHEN Nota >= 4 THEN 'Aprobado'
                        WHEN Nota < 4 THEN 'Desaprobado'
                    END AS Condicion,
                    COUNT(*) AS CantidadAlumnos
                FROM Examen
                GROUP BY Instancia, Condicion;

               """

df_Resultado = pd.read_sql_query(Consulta_SQL, Connection)

df_Resultado
```

### __Subqueries__

```python
#a. Listar los alumnos que en cada instancia obtuvieron una nota mayor al promedio de dicha instancia.

Consulta_SQL = """
                SELECT DISTINCT Nombre, Instancia
                FROM Examen AS E
                WHERE Nota > (SELECT AVG(A.Nota) AS PromedioNota
                              FROM Examen AS A
                              WHERE E.Instancia = A.Instancia)

               """


df_Resultado = pd.read_sql_query(Consulta_SQL, Connection)

df_Resultado
```

```python
# b. Listar los alumnos que en cada instancia obtuvieron la mayor nota de dicha instancia.

Consulta_SQL = """
                SELECT E.Nombre
                FROM Examen AS E
                WHERE E.Nota = (SELECT MAX(A.Nota) AS MaximaNota
                                FROM Examen AS A
                                WHERE E.Instancia = A.Instancia);
                
               """

df_Resultado = pd.read_sql_query(Consulta_SQL, Connection)

df_Resultado
```

```python
Consulta_SQL = """
                  SELECT DISTINCT Instancia, MAX(A.Nota) AS MaximaNota
                  FROM Examen AS A
                  GROUP BY Instancia
                
               """

df_Resultado = pd.read_sql_query(Consulta_SQL, Connection)

df_Resultado
```

```python
# c. Listar el nombre, instancia y nota sólo de los estudiantes que no rindieron ningún Recuperatorio.

Consulta_SQL = """
                SELECT DISTINCT E.Nombre, E.Instancia, E.Nota
                FROM Examen as E
                WHERE E.Nombre NOT IN (SELECT DISTINCT A.Nombre
                                       FROM Examen as A
                                       WHERE A.Instancia LIKE 'Recuperatorio%')

               """

df_Resultado = pd.read_sql_query(Consulta_SQL, Connection)

df_Resultado
```
